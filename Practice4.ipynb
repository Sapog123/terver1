## Определения

Несовместные события
$$P(AB) = 0$$

Независимые события
$$P(AB) = P(A)P(B)$$

(Вообще говоря, привычная "независимость" и математическая "независисмость" - не одно и то же)

Совместная независимость

$$P(ABC) = P(A)P(B)P(C)$$

Попарная независимость A, B, C
$$P(AB) = P(A)P(B)$$ $$P(BC) = P(B)P(C)$$$$P(CA) = P(C)P(A)$$

**Из совместной независимости следует попарная, но из попарной не следует совместная**

**ФОРМУЛА БАЙЕСА**
$$P(A|B) = \frac{P(B|A)P(A)}{P(B)}$$

Формула полной вероятности
$$P(A) = \sum_{i=1}^{n}P(A|\mathcal{H}_i)$$
